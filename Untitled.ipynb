Q1. A company conducted a survey of its employees and found that 70% of the employees use the
company's health insurance plan, while 40% of the employees who use the plan are smokers. What is the
probability that an employee is a smoker given that he/she uses the health insurance plan?

```python
# Define the probabilities
probability_uses_insurance = 0.70  # Probability that an employee uses insurance
probability_smoker_given_uses_insurance = 0.40  # Probability that an employee is a smoker given they use insurance

# Calculate the conditional probability
conditional_probability = probability_smoker_given_uses_insurance / probability_uses_insurance

# Print the result
print(f"The probability that an employee is a smoker given that they use insurance is: {conditional_probability:.2f}")
```

Q3. How does Bernoulli Naive Bayes handle missing values?

Bernoulli Naive Bayes is a variation of the Naive Bayes algorithm that is typically used for binary classification tasks, where the features are binary variables (i.e., they take on only two values, typically 0 and 1). This algorithm is commonly used in text classification and spam detection.

Handling missing values in Bernoulli Naive Bayes can be straightforward because it assumes that features are binary, meaning they are either present (1) or absent (0). When you encounter a missing value for a feature, you can treat it as if the feature is absent (0) and continue with the classification process. This assumption aligns with the binary nature of the Bernoulli distribution.

Here's a step-by-step explanation of how Bernoulli Naive Bayes handles missing values:

Training Phase:

During the training phase, the algorithm calculates probabilities for each feature being 0 or 1 in each class.
If a data point in the training set has a missing value for a particular feature, the algorithm essentially ignores that feature for that data point and continues with the other features.
The algorithm still calculates probabilities for the other features in the class, and these probabilities contribute to the class conditional probabilities used in the Naive Bayes classification formula.
Testing/Classification Phase:

When classifying a new data point with missing values, the algorithm applies Bayes' theorem to compute the probability of the data point belonging to each class.
For features with missing values, the algorithm assumes that they are absent (0) and includes this assumption in the calculation.
